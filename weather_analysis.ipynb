```python
distw = getattr(stats, 'weibull_min')
parametersw = distw.fit(ws)
print(parametersw)
```

```python
# ajustar una distribución de Weibull mínima (weibull_min) a los datos de ws
# entrega los parametros:
# c → Parámetro de forma (shape parameter).
# loc → Parámetro de ubicación (location parameter).
# scale → Parámetro de escala (scale parameter).distw = getattr(stats, 'weibull_min')
parametersw = distw.fit(ws)
```

```python
forma = parametersw[0]
locw = parametersw[1]
scalew =parametersw[2]
```

```python
#Ordenar de menor a mayor para crear las gráficas
x = np.sort(ws) # equivalente al comando seq en R

# Graficando Función de Densidad de Probibilidad con Python
dweibull = stats.weibull_min(forma,locw,scalew).pdf(x) # FDP
```

```python
# gráfica de densidad

plt.hist(ws, intervalos ,color='skyblue',density=1,label='data real')
plt.plot(x, dweibull, label='distribución weibull')

plt.title('Función de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.legend(loc=1)
plt.ylim(0,0.2)
```

```python
# calcular la función de distribución acumulada (CDF, Cumulative Distribution Function) en el valor 8
# Calcula la probabilidad de que una variable aleatoria Weibull sea menor o igual a 8.
stats.weibull_min(forma,locw,scalew).cdf(8)
```

```python
# calcular la probabilidad de que una variable aleatoria Weibull sea mayor que 4
1-stats.weibull_min(forma,locw,scalew).cdf(4)
```

```python
# gráfica de densidad

pweibull = stats.weibull_min(forma,locw,scalew).cdf(x)

g = plt.hist(ws, intervalos ,color='skyblue',density=1,label='data real',cumulative=1)
plt.plot(x, pweibull, label='distribución weibull')
plt.plot(x, 1- pweibull, label='confiabilidad')

plt.title('Función de Densidad de Probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.legend(loc='best')
```

```python
# calcula cuartiles (25, 50, 75%).
# Definicion de las funciones para calcular los cuartiles
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.median()

def q3(x):
    return x.quantile(0.75)
```

```python
# Calculate the hourly quartiles and mean 
hourly_sst = df1.Temperatura.groupby(df1.hour).agg([q1, q2, q3, np.mean])
# Quick plot
hourly_sst.plot(title='BAQ Quartiles horarios y promedio 2015_2020', ylabel="Temperatura °C");
```
